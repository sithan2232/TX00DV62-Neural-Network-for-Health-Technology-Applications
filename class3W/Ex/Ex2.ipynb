```python
%pylab inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import *
from sklearn.model_selection import train_test_split
```

```python
filename=r'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
df=pd.read_csv(filename,index_col=None,header=None,na_values='?')
df=df.dropna()
df.head()
```

```python
df
```

```python
data=df.loc[:,0:12]
labels=1.0*(df.loc[:,13]>0)
print(data.shape,labels.shape)
```

```python
predicted_output=1.0*(data[0]>60)
```

```python
print(confusion_matrix(labels,predicted_output))
#This is using only age(colume[0]) to predict the result. 
```

```python
print(confusion_matrix(labels,predicted_output).T)
```

```python
print(classification_report(labels,predicted_output))
```

```python
from sklearn.linear_model import LinearRegression
 
reg=LinearRegression().fit(data,labels)
predicted_output=reg.predict(data)>0.5
cm=confusion_matrix(labels,predicted_output)
print(cm)
```
